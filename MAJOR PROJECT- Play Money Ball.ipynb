```python
#import all the necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read the dataset

df = pd.read_csv("C:/Users/admin/Desktop/ML/baseball.csv")
df.head()
```

```python
#RA stands for runs allowed.
#RS stand for runs scored.
#OBP stands for On Base Percentage.
#SLG stand for Slugging Percentage.
#BA is Batting Average.
#OOBP is opponent’s OBP.
#OSLG is opponent’s SLG.
#W stands for number of wins in that season.
```

```python
#add a new column, RD where its value is the difference between RS and RA

rd=df['RS']-df['RA']
df["RD"]=rd
df
```

```python
#Display Histograms for the necessary required features

df.hist(column="RS")
plt.title('Runs Scored')

df.hist(column="RA")
plt.title('Runs Allowed')

df.hist(column="OBP")
plt.title(' On Base Percentage')

df.hist(column="SLG")
plt.title(' Slugging Percentage')
```

```python
#slice up the data,into - before the year 2002 and after the year 2002

moneyball0=df[df.Year<2002]
moneyball1=df[df.Year>2002]
```

```python
# Plotting scatterplot

#A Team requires at least 99 wins in order to make the playoffs in 2002. Lets try to visualise this insight.

plt.figure(figsize=(10,8))
plt.scatter(x=moneyball0.W,y = moneyball0.RS,c="red")
plt.scatter(x=moneyball1.W,y = moneyball1.RS,c="blue")
plt.xlabel("Wins")
plt.ylabel("Runs Scored")
plt.axvline(x = 99)
```

```python
#From the above plot we see the estimation of 99 wins makes sense ,
#since there appears to be 40 observations (red) who haven’t made it to the playoffs with ≥ 99 wins.
```

```python
#coming up with the relation between RD and Wins , to get a team into the postseason

x = np.array(df.RD)
y = np.array(df.W)

# Deriving slope,intercept values

slope, intercept = np.polyfit(x, y, 1)
abline_values = [slope * i + intercept for i in x]

#Plotting the figure

plt.figure(figsize=(10,8))
plt.scatter(x, y)
plt.plot(x, abline_values, 'b')
plt.title("Slope = %s" % (slope))
plt.xlabel("Run Difference")
plt.ylabel("Wins")
plt.show()
```

```python
#We can clearly see that the relationship between the two variables is linear

print(np.corrcoef(x,y))
```

```python
#therefore, there is a strong correlation of 0.9385 between RD and number of wins.
```
