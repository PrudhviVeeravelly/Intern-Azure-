```python
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=sb.load_dataset('planets')
set(df.distance)
```

```python
df.distance.replace(np.NaN,0)  #replacing nan values with 0
```

# HISTOGRAM

```python
sb.distplot(df.distance.replace(np.NaN,0),kde=False)  #to display histogram , kde is display the plot line
```

```python
sb.set_style('white',{'axes.facecolor':'yellow'})
sb.distplot(df.distance.replace(np.NaN,0),kde=False) #for displaying kde graph
```

```python
sb.set_style('white',{'axes.facecolor':'yellow'})
sb.distplot(df.distance.replace(np.NaN,0),hist=False) #for displaying histogram graph
```

# HEATMAP-iris dataset

```python
da=sb.load_dataset('iris')
d=da.drop('species',axis=1)
sb.heatmap(d)
```

```python
sb.pairplot(da,hue='species',kind='scatter',diag_kind='kde',palette='rainbow')
#diag_kind is for displaying the diagonal graphs
#kind ='scatter' for scattering of data points
```
