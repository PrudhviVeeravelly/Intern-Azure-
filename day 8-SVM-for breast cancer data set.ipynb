```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
```

```python
bc=datasets.load_breast_cancer()
x=bc.data[:,1:3] #we want to take only the first 2 features
y=bc.target
x.shape
```

```python
C=1.0 #SVM regulation parameter
svc=SVC(kernel='rbf',C=1,gamma=1).fit(x,y) #replace kernel with linear ,rbf , poly , sigmoid ,precomputed
```

```python
#create a mesh to plot in

x_min,x_max=x[:,0].min()-1,x[:,0].max()+1
y_min,y_max=x[:,1].min()-1,x[:,1].max()+1
h=(x_max/x_min)/100
xx,yy=np.meshgrid(np.arange(x_min,x_max,h), np.arange(y_min,y_max,h))
```

```python
plt.subplot(1,1,1)
Z=svc.predict(np.c_[xx.ravel(),yy.ravel()]) #ravel is used to export the complete data in a row or column
                                            #.c_ is used for getting data in pairs
Z=Z.reshape(xx.shape)
plt.contourf(xx,yy,Z,cmap=plt.cm.Paired,alpha=0.2)


plt.scatter(x[:,0],x[:,1],c=y,cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(),xx.max())
plt.title('SVC with linear kernel')
plt.show()
```
