```python
import pandas as pd
import numpy as np
data=pd.read_csv("C:/Users/admin/Desktop/ML/data/train.csv")
ip=np.array(data.loc[:,'PassengerId']).reshape(-1,1)
op=np.array(data.loc[:,'Survived']).reshape(-1,1)
from sklearn.model_selection import train_test_split
xtr,xte,ytr,yte=train_test_split(ip,op,test_size=0.2)

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtr,ytr)
lr.predict([[200]])
#print(metrics.accuracy_score(yte,lr.predict(xte)))
import matplotlib.pyplot as plt
plt.scatter(xte,yte,color='red')
plt.plot(xte,lr.predict(xte),color='black',linewidth=1)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
ip=np.array(data.loc[:,'PassengerId']).reshape(-1,1)
op=np.array(data.loc[:,'Survived']).reshape(-1,1)
knn=KNeighborsClassifier(1)
knn.fit(xtr,ytr)
from sklearn.model_selection import train_test_split
xtr,xte,ytr,yte=train_test_split(ip,op,test_size=0.2)
from sklearn import metrics
print(metrics.accuracy_score(yte,knn.predict(xte))*100)
knn.predict([[2]])

import matplotlib.pyplot as plt
plt.scatter(xte,yte,color='red')
plt.plot(xte,knn.predict(xte),color='black',linewidth=1)
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
ip=np.array(data.loc[:,'PassengerId']).reshape(-1,1)
op=np.array(data.loc[:,'Survived']).reshape(-1,1)
nb=GaussianNB()
nb.fit(xtr,ytr)
from sklearn.model_selection import train_test_split
xtr,xte,ytr,yte=train_test_split(ip,op,test_size=0.2)
print(metrics.accuracy_score(yte,nb.predict(xte))*100)
print(nb.predict([[2]]))

import matplotlib.pyplot as plt
plt.scatter(xte,yte,color='red')
plt.plot(xte,nb.predict(xte),color='black',linewidth=1)
plt.show()
```

```python
from sklearn import tree
dt=tree.DecisionTreeClassifier()
ip=np.array(data.loc[:,'PassengerId']).reshape(-1,1)
op=np.array(data.loc[:,'Survived']).reshape(-1,1)
dt.fit(xtr,ytr)
from sklearn.model_selection import train_test_split
xtr,xte,ytr,yte=train_test_split(ip,op,test_size=0.2)
print(metrics.accuracy_score(yte,dt.predict(xte))*100)
print(dt.predict([[2]]))

import matplotlib.pyplot as plt
plt.scatter(xte,yte,color='red')
plt.plot(xte,dt.predict(xte),color='black',linewidth=1)
plt.show()
```
