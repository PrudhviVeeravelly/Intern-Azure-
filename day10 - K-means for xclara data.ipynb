```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
d=pd.read_csv('C:/Users/admin/Desktop/ML/xclara.csv')
d
```

```python
#number of clusters
k=4
kmean=KMeans(n_clusters=k)  #import model
```

```python
#train the data

kmean = kmean.fit(d)  #train the model
```

```python
labels=kmean.labels_  #array conversion of the predicted values of our data into categories

set(labels) #the set of those unique labels (here we get 4 becuase centroid value is 4 )
```

```python
centroids=kmean.cluster_centers_

centroids.shape
```

```python
x_test=[[49.6,45],[56,90.4],[33.33,69]]
```

```python
prediction= kmean.predict(x_test)
prediction
```

```python
colors=['blue','black','yellow','red']
```

```python
y=0
for x in labels:
    plt.scatter(d.iloc[y,0],d.iloc[y,1],color=colors[x])   #y is for accessing every row in the 2 columns(V1,V2 in the data)
    y=y+1
for x in range(k):
    lines=plt.plot(centroids[x,0],centroids[x,1],'kx') #kx is like define variable where u have marker and color,
                                                        #more variables can be found in help
    plt.setp(lines,ms=15.0)
    plt.setp(lines,mew=2.0)

plt.xlabel('v1')
plt.ylabel('v2')
plt.title(' Number of clusters k= {} '.format(k) )
plt.show()
```
