```python
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder
```

```python
import pandas as pd

dataset = [['Milk','Onion','Nutmeg','Kidney Beans','Eggs','Yogurt'],
           ['Dill','Onion','Nutmeg','Kidney Beans','Eggs','Yogurt'],
           ['Milk','Apple','Kidney Beans','Eggs'],
           ['Milk','Unicorn','Corn','Kidney Beans','Yogurt'],
           ['Corn','Onion','Kideny Beans','Ice cream','Eggs']]

'''
dataset = [['Milk','Butter','Bread','Eggs'],
           ['Onion','Potato','Eggs','Salt'],
           ['Milk','Bread','Potato','Salt'],
           ['Milk','Bread','Potato','Salt','Onion'],
           ['Butter','Bread','Eggs','Milk']]
'''
```

```python
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)  #te_ary will contain all the values as True or False
df=pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
frequent_itemsets=apriori(df,min_support=0.5,use_colnames=True)
frequent_itemsets
```

```python
#to add an new olumn, and should contain the length of the item set - to calculate the confidence

frequent_itemsets['length']=frequent_itemsets['itemsets'].apply(lambda x:len(x)) #lambda is used to create a function
frequent_itemsets
```

```python
frequent_itemsets[(frequent_itemsets['support']>=0.6) & (frequent_itemsets['length']==2)]
```
