```python
import pandas as pd
import numpy as np
from sklearn import datasets
d=datasets.load_boston()
data=pd.DataFrame(d.data,columns=d.feature_names)
data1=pd.DataFrame(d.target,columns=['ABC'])
inp=np.array(data.loc[:,'INDUS']).reshape(-1,1)
op=np.array(data1.loc[:,'ABC']).reshape(-1,1)


from sklearn.model_selection import train_test_split
xtr,xte,ytr,yte=train_test_split(inp,op,test_size=0.2)

from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=5)
l=poly.fit_transform(xtr)
poly.fit(l,ytr)

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(l,ytr)
#print(lr.predict([[20]]))

import matplotlib.pyplot as plt
plt.scatter(inp,op,color='red')
plt.plot(xte,lr.predict(poly.fit_transform(xte)),color='black',linewidth=1)
plt.show()
```
